```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("air_bnb.csv")
```

```python
df.head()
```

```python
df.columns
```

# checking missing values 

```python
print(df.isnull().sum())
```

```python
df.info()
```

# handling missing values 

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review':df['last review'].min()},inplace = True)
```

```python
df.dropna(subset=['NAME','host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['license','house_rules'],errors = 'ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace = True)
df.info()
```

# Descriptive Statistics 

```python
df.describe()
```

# Visualization

# what is the distribution listing prices ?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50, kde = True,color='red')
plt.title('Distribution of listing price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

```python
This Histogram shows fairly even distribution of 
listing prices across different price ranges, indicating no
particular concentration of listings in any specific price range.
The KDE line helps visualize the even spread more clearly,
confirming that the dataset contains listings with a wide variety of prices.
```

# How are room type distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='room type',data=df,color='hotpink')
plt.title(' Room type ditribution ')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

# How are listings distributed across different neighbourhood?

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group',data = df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by neighbourhood group')
plt.xlabel('count')
plt.ylabel('neighbourhood group')
plt.show()
```

# What is the relationship between price & roomtype?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title('Price VS Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

# How has the number of reviews change over with time

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size() 
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show() 
```
